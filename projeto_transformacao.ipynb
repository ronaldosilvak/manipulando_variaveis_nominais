```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
# json_normalize -> essencial para transformar dados semiestruturados no formato json em dataframes
dados = pd.json_normalize(dados['info_moveis'])
dados
```

```python
dados.info()
```

```python
colunas = list(dados.columns)
colunas
```

```python
# transforma cada elemento presente em uma lista, separando-os e colocando cada um em uma linha da tabela
dados = dados.explode(colunas[3:])
dados
```

```python
# drop -> joga o index antigo fora, assim mantemos apenas o novo. Desta forma, evitando dois indÃ­ces na tabela
dados.reset_index(inplace=True, drop=True)
dados
```

```python
dados.info()
```

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype('int64')
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype('int64')
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

```python
dados['preco']
```

```python
# strip --> retira os espaÃ§os em branco do inÃ­cio e do fim
# replace ('valor que vai sair', 'valor que vai entrar no lugar')
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

```python
dados['descricao_local'][3169]
```

```python
# ^ operador de negaÃ§Ã£o -> assim estamos negando ^a-zA-Z0-9 esse script, a-z (minÃºsculo) A-Z (maiÃºsculo) de 0 a 9. E estamos aceitando os caracteres \-\' (- e ')
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
# (?<!\w)-(?!\w), aqui o script faz uma verificaÃ§Ã£o antes e depois do hÃ­fen (-), onde o mesmo serÃ¡ removido apenas se NÃƒO fazer parte de uma palavra composta, ou seja,
# nÃ£o possuir palavra antes ou depois dele
dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
